## Extração dos textos das imagens

```python
import pytesseract
from PIL import Image
import os

pasta_imagens = 'Systematic'

with open('Dados.txt', 'w', encoding='utf-8') as arquivo_txt:
    for arquivo in os.listdir(pasta_imagens):
        if arquivo.endswith('.png') or arquivo.endswith('.jpg') or arquivo.endswith('.jpeg'):
            caminho_imagem = os.path.join(pasta_imagens, arquivo)

            img = Image.open(caminho_imagem)

            texto = pytesseract.image_to_string(img)

            arquivo_txt.write(f"Texto extraído da {arquivo}:\n{texto}\n")
            arquivo_txt.write("-" * 40 + "\n\n")
```

```python
from transformers import MarianMTModel, MarianTokenizer

# Novo nome do modelo
model_name = 'Helsinki-NLP/opus-mt-en-ROMANCE'

# Carregar o modelo e o tokenizer
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Função para tradução
def traduzir_texto(texto):
    tokens = tokenizer(texto, return_tensors="pt", padding=True, truncation=True)
    traducao = model.generate(**tokens)
    texto_traduzido = tokenizer.batch_decode(traducao, skip_special_tokens=True)
    return texto_traduzido[0]

# Carregar o documento txt
with open('Dados.txt', 'r', encoding='utf-8') as f:
    texto = f.read()

# Traduzir o texto
texto_traduzido = traduzir_texto(texto)

# Salvar a tradução
with open('Dados_traduzidos.txt', 'w', encoding='utf-8') as f:
    f.write(texto_traduzido)

print("Tradução concluída e salva em Dados_traduzidos.txt.")
```

## Rascunho - Tradução Automática com Google Translator

```python
from deep_translator import GoogleTranslator

# Função para ler o arquivo "Dados.txt"
def ler_arquivo(arquivo):
    with open(arquivo, 'r', encoding='utf-8') as f:
        conteudo = f.read()
    return conteudo

# Função para traduzir o conteúdo do arquivo
def traduzir_conteudo(conteudo, idioma_destino='pt'):
    tradutor = GoogleTranslator(source='auto', target=idioma_destino)
    return tradutor.translate(conteudo)

# Função para salvar o conteúdo traduzido no arquivo "Dados_pt.txt"
def salvar_arquivo(conteudo, arquivo_destino):
    with open(arquivo_destino, 'w', encoding='utf-8') as f:
        f.write(conteudo)

# Função principal para executar a tradução do arquivo
def traduzir_arquivo():
    arquivo_origem = 'Dados.txt'
    arquivo_destino = 'Dados_pt.txt'

    # Lendo o conteúdo do arquivo original
    conteudo = ler_arquivo(arquivo_origem)

    # Traduzindo o conteúdo
    conteudo_traduzido = traduzir_conteudo(conteudo)

    # Salvando o conteúdo traduzido em um novo arquivo
    salvar_arquivo(conteudo_traduzido, arquivo_destino)
    print(f'Tradução concluída! O conteúdo traduzido foi salvo em {arquivo_destino}')

# Executando a função principal
if __name__ == '__main__':
    traduzir_arquivo()
```
